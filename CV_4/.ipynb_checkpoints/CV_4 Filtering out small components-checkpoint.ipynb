<h5>Filtering out small components</h5>

```python
def filter_out_small_components(thresh_img, connectivity, min_pixels):

    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(thresh_img, connectivity=connectivity)

    # Filter out components with fewer than 100 pixels
    #stats[i, cv2.CC_STAT_AREA] gets the i'th area of the connected component i, it's the same as 
    # stats[i, 4]
    min_pixels = min_pixels
    for i in range(1, num_labels):
        if stats[i, cv2.CC_STAT_AREA] < min_pixels:
            labels[labels == i] = 0

    # Convert labeled image back to binary
    filtered_image = np.uint8(labels > 0) * 255
    return filtered_image
```

```python
th1_filtered = filter_out_small_components(th1, 8, 100)
plt.imshow(th1_filtered ,cmap='gray')
```
