```python
def getDistance(y_1,x_1,y_2,x_2):
    distance = math.sqrt(math.pow((y_2-y_1),2) + math.pow((x_2-x_1),2))
    return distance
```

```python
def getN4(y,x):
    n4 = []
    try: 
        n4.append([y-1,x])
    except:
        print("No Top")

    try: 
        n4.append([y+1,x])
    except:
        print("No bottom")

    try: 
        n4.append([y, x-1])
    except:
        print("No left")

    try: 
        n4.append([y, x+1])
    except:
        print("No right")
    return n4
```

```python
def getPath(img,y_1,x_1,y_2,x_2):
    visited = []
    pos_y = y_1
    pos_x = x_1
    while (pos_y != y_2) or (pos_x != x_2):
        champ = []
        champ_dist = 10000
        pos_n4 = getN4(pos_y,pos_x)
        valid_pos = []
        for pos in pos_n4:
            if img[pos[0],pos[1]] == 0 and pos not in visited:
                valid_pos.append(pos)
        for pos in valid_pos:
            if getDistance(pos[0],pos[1],y_2,x_2) <= champ_dist:
                champ_dist = getDistance(pos[0],pos[1],y_2,x_2)
                champ = pos
        pos_y = champ[0]
        pos_x = champ[1]
        visited.append([pos_y,pos_x])
        print([pos_y,pos_x])
    return visited
        
```
