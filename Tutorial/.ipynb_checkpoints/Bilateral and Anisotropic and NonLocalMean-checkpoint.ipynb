```python
import cv2
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import cv2.ximgproc
```

<h3>Params for bilateralFilter()</h3>
src: the input image to be filtered

d: diameter of each pixel neighborhood used during filtering

sigmaColor: parameter that controls the degree of color similarity between the neighboring pixels

sigmaSpace: parameter that controls the degree of spatial proximity between the neighboring pixels

```python
img = cv2.imread("lenna.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

<h5>Experimenting with different noise, before using the denoisers</h5>

```python
import random

def add_noise(img):
    row, col, channels = img.shape
    
    num_of_pixels = random.randint(300,10000)
    for i in range(num_of_pixels):
        
        y = random.randint(0, row-1)
        x = random.randint(0, col-1)
        img[y][x] = 255
        
    num_of_pixels = random.randint(300,10000)
    for i in range(num_of_pixels):
        
        y = random.randint(0, row-1)
        x = random.randint(0, col-1)
        img[y][x] = 0
        
        
    return img
```

```python
#img = add_noise(img)
gauss = np.random.normal(0,1,img.size)
#gauss array in img shape shapen. Die 255 sind die negativen
gauss = gauss.reshape(img.shape[0], img.shape[1],img.shape[2]).astype('uint8')
#Add Gaussian noise to the image
img = cv2.add(img,gauss)
```

```python
denoised_image_gauss = cv2.GaussianBlur(img, (5,5), 5)
plt.imshow(denoised_image_gauss)
```

```python
plt.imshow(img)
```

```python
#this is far and away more effective against salt and pepper noise, because the way that bilateral
#weighs the intensities of its neighboring pixels, doesn't work well with extremes like 0 and 255
median = cv2.medianBlur(img,5)
plt.imshow(median)
```

<h2>Params for Anisotropic Diffusion</h2>


src: The input image to be filtered. The size and content of this image will directly affect the output of the function.

dst: The output image that will contain the result of the filtering. This image will have the same size and type as the input image.

alpha: The diffusion coefficient. This parameter controls the amount of smoothing that is applied to the image. A smaller value of alpha will result in less smoothing, while a larger value of alpha will result in more smoothing.

kappa: The gradient threshold. This parameter controls the strength of the edges in the image. A smaller value of kappa will allow weaker edges to be smoothed more, while a larger value of kappa will preserve the edges in the image.

niters: The number of iterations of the anisotropic diffusion filtering algorithm. This parameter controls how many times the smoothing operation is applied to the image. A larger number of iterations will result in a more heavily smoothed image.

```python
bilateral = cv2.bilateralFilter(img, 5, 20, 50)
plt.imshow(bilateral)
```

```python

# Apply anisotropic diffusion with appropriate parameters
filtered = cv2.ximgproc.anisotropicDiffusion(img,0.05, 50, 6)

# Display the original and filtered images side by side
plt.imshow(filtered)

```

<h3>Non Local Mean</h3>

src: The input color image to be denoised.

dst: The output denoised image. This should have the same size and type as the input image.

h: The parameter controlling the amount of denoising. Larger values of h will result in stronger denoising, while smaller values will result in less denoising. The optimal value of h depends on the amount of noise in the image.

hForColorComponents: A vector of parameters controlling the amount of denoising for each color component. By default, this is set to [h/2, h/2, h/2], which means that each color component is treated equally.

templateWindowSize: The size of the window used for finding similar patches in the image. Larger values of templateWindowSize will result in more global denoising, while smaller values will result in more local denoising.

searchWindowSize: The size of the window used for searching for similar patches in the image. Larger values of searchWindowSize will result in more smoothing, while smaller values will result in less smoothing.

in this case: (src, output, h, hForColored, templateWindowSize, searchWindowSize)

```python
dst = cv2.fastNlMeansDenoisingColored(img, None, 15, 10, 5, 20)
plt.imshow(dst)
```
