<h2>K-Means++ and Fuzzy C-Means</h2>

```python
import cv2
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import ndimage
from sklearn.cluster import KMeans
import skfuzzy as fuzz
```

```python
import cv2
import numpy as np
from sklearn.cluster import KMeans

# Load the image
gray = cv2.imread('circles_rauschen.png', cv2.IMREAD_GRAYSCALE)

# Initialize the KMeans++ model
kmeans = KMeans(n_clusters=3, init='k-means++')

# Reshape the image into a 2D array
data = gray.reshape(-1, 1)

# Fit the model to the data
kmeans.fit(data)

# Get the labels and centers of the clusters
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Threshold the image using the KMeans++ centers
threshold = np.mean(centers)

# Apply the threshold to the image
binary = gray > threshold
```

```python
plt.imshow(gray, cmap='gray')
```

```python
plt.imshow(binary, cmap='gray')
```

<h3>Fuzzy C-Means</h3>

<h4>Ohne Bluring</h4>

The fuzz.cluster.cmeans function takes several arguments:

gray.reshape(-1, 1) converts the grayscale image to a 1D array that can be used as input to the clustering algorithm.

nc is the number of clusters.

2 is the fuzziness exponent.

error is the maximum error between consecutive iterations.

maxiter is the maximum number of iterations.

init is the initial cluster centers. If set to None, the algorithm will randomly initialize the centers.

```python
# Load the image
gray = cv2.imread('circles_rauschen.png', cv2.IMREAD_GRAYSCALE)

# Apply Fuzzy C-Means clustering
nc = 3 # number of clusters
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(gray.reshape(-1, 1), nc, 2,
                                                 error=0.005, maxiter=1000, init=None)

# Threshold the image using the cluster centers
threshold = np.mean(cntr)
binary = gray > threshold

plt.imshow(binary, cmap='gray')
```

<h4>Mit Bluring</h4>

```python
# Load the image
gray = cv2.imread('circles_rauschen.png', cv2.IMREAD_GRAYSCALE)
gray = cv2.medianBlur(gray,5)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
# Apply Fuzzy C-Means clustering
nc = 3 # number of clusters
cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(gray.reshape(-1, 1), nc, 2,
                                                 error=0.005, maxiter=1000, init=None)

# Threshold the image using the cluster centers
threshold = np.mean(cntr)
binary = gray > threshold

plt.imshow(binary, cmap='gray')
```
