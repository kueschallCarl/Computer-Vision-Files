```python
import cv2
import numpy as np

def nothing(x):
    pass

cap = cv2.VideoCapture(0, cv2.CAP_DSHOW);

cv2.namedWindow("Tracking")
cv2.createTrackbar("LH", "Tracking", 0, 255, nothing)
cv2.createTrackbar("LS", "Tracking", 0, 255, nothing) 
cv2.createTrackbar("LV", "Tracking", 0, 255, nothing)
cv2.createTrackbar("UH", "Tracking", 255, 255, nothing)
cv2.createTrackbar("US", "Tracking", 255, 255, nothing)
cv2.createTrackbar("UV", "Tracking", 255, 255, nothing)


cv2.namedWindow("Tracking2")
cv2.createTrackbar("LH2", "Tracking2", 0, 255, nothing)
cv2.createTrackbar("LS2", "Tracking2", 0, 255, nothing)
cv2.createTrackbar("LV2", "Tracking2", 0, 255, nothing)
cv2.createTrackbar("UH2", "Tracking2", 255, 255, nothing)
cv2.createTrackbar("US2", "Tracking2", 255, 255, nothing)
cv2.createTrackbar("UV2", "Tracking2", 255, 255, nothing)



while True:
    #frame = cv2.imread('smarties.jpg')
    _, frame = cap.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    l_h = cv2.getTrackbarPos("LH", "Tracking")
    l_s = cv2.getTrackbarPos("LS", "Tracking")
    l_v = cv2.getTrackbarPos("LV", "Tracking")
    
    u_h = cv2.getTrackbarPos("UH", "Tracking")
    u_s = cv2.getTrackbarPos("US", "Tracking")
    u_v = cv2.getTrackbarPos("UV", "Tracking")
    
    l_b = np.array([l_h, l_s, l_v])
    u_b = np.array([u_h, u_s, u_v])
    
    mask = cv2.inRange(hsv, l_b, u_b)
    
    l_h2 = cv2.getTrackbarPos("LH2", "Tracking2")
    l_s2 = cv2.getTrackbarPos("LS2", "Tracking2")
    l_v2 = cv2.getTrackbarPos("LV2", "Tracking2")
    
    u_h2 = cv2.getTrackbarPos("UH2", "Tracking2")
    u_s2 = cv2.getTrackbarPos("US2", "Tracking2")
    u_v2 = cv2.getTrackbarPos("UV2", "Tracking2")
    
    l_b2 = np.array([l_h2, l_s2, l_v2])
    u_b2 = np.array([u_h2, u_s2, u_v2])
    
    mask2 = cv2.inRange(hsv, l_b2, u_b2)
    
    res1 = cv2.bitwise_and(frame, frame, mask=mask)
    res2 = cv2.bitwise_and(frame, frame, mask=mask2)
    
    
    res = cv2.addWeighted(res1,0.5,res2,0.5,0)
    cv2.imshow("frame",frame)
    cv2.imshow("mask",mask)
    cv2.imshow("mask2",mask2)
    cv2.imshow("res",res)
    
    
    key = cv2.waitKey(1)
    if key ==27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
