<h2>Gaussian Mixture Thresholding</h2>

Important: More than 2 components don't really work here, because we're getting the mean of the Erwartungswert of the gaussians, to seperate foreground and background. If we choose 3 components, then that mean is trying to differentiate more than just fore and background, which doesn't make any sense.

```python
import cv2
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import ndimage
from sklearn.mixture import GaussianMixture
```

```python
# Load the noisy image
img = cv2.imread('circles_rauschen.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(img,cmap='gray')
```

```python
# Reshape the image to a 1D array
img_1d = np.reshape(img, (img.shape[0] * img.shape[1], 1))

# Fit a Gaussian Mixture Model with two components to the image
gmm = GaussianMixture(n_components=2)
gmm.fit(img_1d)

# Get the mean and standard deviation of the two Gaussian distributions
means = gmm.means_
stds = np.sqrt(gmm.covariances_)

# Set the threshold value as the average of the n means
threshold = np.mean(means)

# Apply thresholding to the image
binary_img = np.where(img >= threshold, 255, 0)

# Display the final Gaussians
plt.hist(img_1d, bins=256, density=True, alpha=0.4, color='g')
x = np.linspace(0, 255, 256)
for i in range(len(means)):
    y = 1/(stds[i] * np.sqrt(2 * np.pi)) * np.exp(-(x - means[i])**2 / (2 * stds[i]**2))
    plt.plot(x, y.reshape(-1,), linewidth=2)

plt.title('Final Gaussian Mixtures')
plt.show()
```
