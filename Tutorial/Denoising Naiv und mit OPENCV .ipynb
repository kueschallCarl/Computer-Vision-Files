<h2>NAIVE IMPLEMENTIERUNG NIX GUT</h2>

```python
def convolve(image, kernel):
    #get spatial dimensions of the image, along with the spatial dimensions of the kernel
    (iH, iW)= image.shape[:2]
    (kH, kW) = kernel.shape[:2]
    #allocate memory for the output image, taking care to "pad" the borders of the input image so the
    #spatial size (width and height) are not reduced
    pad = (kW-1)//2
    image = cv2.copyMakeBorder(image, pad, pad, pad, pad, cv2.BORDER_REPLICATE)
    output = np.zeros((iH,iW), dtype="float32")
    #loop over the input image, "sliding" the kernel across each (x,y)-coordinate from left-to-right
    #and top to bottom
    for y in np.arange(pad, iH + pad):
        for x in np.arange(pad, iW+pad):
            #extract the ROI of the image by extracting the center region of the current px-coords
            roi = image[y-pad:y+pad+1, x- pad:x +pad +1]
            #element-wise multiplication
            k = (roi * kernel).sum()
            #store the convolved value in the output (x,y)-coord of the output image
            output[y-pad, x-pad] = k
    #rescale the output image to be in color range
    output = rescale_intensity(output, in_range=(0,255))
    output = (output*255).astype("uint8")
    
    return output
```

```python
img = cv2.imread('circles_rauschen.png')
plt.imshow(img)
```

```python
kernel = np.array([     [0.077847, 0.123317, 0.077847],
                        [0.123317, 0.195346, 0.123317],
                        [0.077847, 0.123317, 0.077847]
], dtype=np.float32)
```

```python
output = convolve(img, kernel)
```

```python
plt.imshow(output,cmap='gray')
```

```python
"""The line of code roi = image[y-pad:y+pad+1, x-pad:x+pad+1] is used to extract a sub-image or a region
of interest (ROI) from the padded input image around the current pixel being convolved.

The y-pad:y+pad+1 and x-pad:x+pad+1 slicing notation creates a rectangular sub-array
of the input image with height and width equal to the kernel size,
centered around the current pixel being convolved. 
The pad value is used to offset the slicing so that the sub-array is centered on the current pixel.
The +1 value is used to ensure that the sub-array has the correct size when the kernel size is odd.

For example, if the kernel size is 3x3, and we are convolving the pixel at row y=5 and column x=7, 
the slicing notation would be:

image[4:7, 6:9]

This would extract a 3x3 sub-array of the input image centered around the pixel at (5, 7). 
This sub-array is then multiplied element-wise with the kernel and summed to obtain the output value
for the current pixel in the output array."""
```
