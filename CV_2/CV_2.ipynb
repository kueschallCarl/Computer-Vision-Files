```python
def getN2(y,x):
    n2 = []
    n2.append([y-1,x])
    n2.append([y, x-1])
    return n2
```

```python
def twoPass(img):
    #the px with the key label should switch to its value label in the second pass
    second_labels = {}
    #equal array of the img to match the labels to the aquivalent pixels
    labels = np.zeros_like(img)
    #to set the labels background super high, so that labels next to background are always smaller 
    background_label_value = 255
    labels[0==0] = background_label_value
    id = 0
    x=0
    y=-1
    
    #we iterate through each line and each pixel of that line, increasing the y and x respectively
    #and resetting x when switching to a new line
    for line in img:
        y+=1
        print("New Line")
        print("Y: "+ str(y))
        x=-1
        for px in line:
            x+=1
            print("X: "+ str(x))
            if px == 255:
                #getN2 returns coordinates, not values
                n = getN2(y,x)
                print("N of "+"["+str(y)+","+str(x)+"]"+ ":" + str(n))
                n_values = []
                
                #should the current pixel be situated at the edge of the image,
                #the following two cases can occur
                north_only = False
                west_only = False
                
                if n[0][0] >= 0 and n[0][1] >=0:
                    north = img[n[0][0]][n[0][1]]
                    #to prevent background pixels to be one of the neighbors
                    if north != 0:
                        n_values.append(north)
                else:
                    print("For px "+"["+str(y)+","+str(x)+"]"+ ":" + "west_only")
                    west_only = True
                    
                if n[1][0] >= 0 and n[1][1] >=0:
                    west = img[n[1][0]][n[1][1]]
                    #to prevent background pixels to be one of the neighbors
                    if west != 0:
                        n_values.append(west)
                else:
                    print("For px "+"["+str(y)+","+str(x)+"]"+ ":" + "north_only")
                    north_only = True
              
                single = 0
                #if there are no foreground neighbors, increase id, label,
                #and declare a new second label
                if n_values == []:
                    id+=1
                    second_labels[id] = id
                    labels[y,x] = id
                else:
                    if west_only == True:
                        west_label = labels[n[1][0]][n[1][1]]     
                        labels[y,x] = west_label
               
                    elif north_only == True:
                        north_label = labels[n[0][0]][n[0][1]]
                        labels[y,x] = north_label
                        
                    else:
                        north_label = labels[n[0][0]][n[0][1]]
                        west_label = labels[n[1][0]][n[1][1]]
                            
                        if north_label == west_label:
                            labels[y,x] = west_label
                        else:
                            if north_label < west_label:
                                labels[y,x] = north_label
                                second_labels[west_label] = north_label
                            else:
                                labels[y,x] = west_label
                                second_labels[north_label] = west_label
    
    
    #second pass
    y=-1
    for line in labels:
        y+=1
        x=-1
        for px in line:
            x+=1
            if labels[y][x]!=background_label_value:
                labels[y][x] = np.int32(second_labels[labels[y][x]]).copy()
    print(second_labels)
    return labels
                    
            
            
```

```python
test = [[255,255,0,255,255,255,0,255],
           [255,255,0,255,0,255,0,255],
           [255,255,255,255,0,0,0,255],
           [0,0,0,0,0,0,0,255],
           [255,255,255,255,0,255,0,255],
           [0,0,0,255,0,255,0,255],
           [255,255,255,255,0,0,0,255],
           [255,255,255,255,0,255,255,255]]
```
