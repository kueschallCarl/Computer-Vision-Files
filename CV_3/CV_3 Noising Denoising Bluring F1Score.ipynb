<h2>Aufgabe 1</h2>

```python
img = cv2.imread("circles1.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
uniques = np.unique(img_gray, return_counts=True)
distincts = []
for val in range(len(uniques[1])):
    if (uniques[1][val] / np.sum(uniques[1])) > 0.04:
       distincts.append(val)
    
masks = []
for th in distincts:
    img_th = np.zeros_like(img_gray)
    img_th[img_gray == th] = 255
    masks.append(img_th)
```

```python
plt.imshow(masks[3], cmap = 'gray')
```

<h3>Salt and Pepper Noise</h3>

Entscheiden zwischen random Menge an Noise oder fest. Je mehr px versalzen/verpfeffert werden, desto mehr Noise gibts. Obviously

```python
def add_noise(img):
    row, col, channels = img.shape
    
    #num_of_pixels = random.randint(300,10000)
    num_of_pixels = 4000
    
    for i in range(num_of_pixels):
        
        y = random.randint(0, row-1)
        x = random.randint(0, col-1)
        img[y][x] = 255
        
    #num_of_pixels = random.randint(300,10000)
    num_of_pixels = 4000
    for i in range(num_of_pixels):
        
        y = random.randint(0, row-1)
        x = random.randint(0, col-1)
        img[y][x] = 0
        
        
    return img
```

```python
img_salt = add_noise(img)
plt.imshow(img_salt)
```

<h3>Poisson Noise</h3>

Check OneNote for thorough explanation of Poisson distribution in CV

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
noise_low = np.random.Generator(np.random.PCG64()).poisson(lam=5, size=gray.shape)
noise_med = np.random.Generator(np.random.PCG64()).poisson(lam=20, size=gray.shape)
noise_high = np.random.Generator(np.random.PCG64()).poisson(lam=50, size=gray.shape)

# Add the generated noise to the original image
noisy_image_low = gray + noise_low.astype('uint8')
noisy_image_med = gray+ noise_med.astype('uint8')
noisy_image_high = gray + noise_high.astype('uint8')



```

<h3>Speckle Noise</h3>

```python
# Generate speckle noise
speckle = np.random.normal(loc=0, scale=0.001, size=gray.shape)
noisy_image = gray + gray * speckle
# Convert the noisy image to uint8 format
noisy_image = np.uint8(np.clip(noisy_image, 0, 255))
```

```python
plt.imshow(noisy_image, cmap='gray')
```

Fazit: Speckle Noise ist allgemein sehr stark. Scale veraendern hilft. Loc<0 -> Dunkleres Bild und andersherum.

Gaussian Blur

```python
#stell dir die gaussian bell curve vor. Px im Kernel die weiter weg vom betrachteten px sind, bekommen
#weniger gewicht (0.184343 fuer px(y+3,x+5), 0.83434 fuer px(y+1,x-1)). Wie viel gewicht sie bekommen
#ist von der Gauss Bell Curve festgelegt. Die X-Achse ist also wie die Distanz vom original px, wobei
#x0 der hochpunkt der Bell Curve ist.

#ist generell besser als averaging wegen der gewichtung, am besten gegen Gaussian noise
denoised_image_gauss = cv2.GaussianBlur(img, (5, 5), 0)
plt.imshow(denoised_image_gauss)
```

Median Blur

```python
#Take all intensities of the kernel, sort them in ascending order.
#Take the mean(value in the middle of the array), that mean becomes the new intensity for that px

#ist schlecht bei extremem Noise. Man nehme an ein kernel ist voll von falschen pixeln, da ist 
#die wahrscheinlichkeit hoch, dass ein falscher pixel genommen wird
#cv2.medianBlur(src, kernel_size)
denoised_image_median = cv2.medianBlur(img,21)
plt.imshow(denoised_image_median)
```

<h2>F1 Score</h2>

ich brauch masken vom wahren bild und vom bearbeiteten bild. In dem Fall vom denoised bild.
Ich erhalte die Masken vom denoised mit opencv thresholding, anders als oben in Aufgabe 1, um hier beide gleich zu thresholden. Dann werden die binaeren werte px fuer px verglichen pro maske. Recall und precision werden berechnet und dann in die f1_score formel eingetragen.

```python
masks_raw = masks
img = cv2.imread("circles1.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_denoised = denoised_image_gauss
img_denoised = cv2.cvtColor(img_denoised, cv2.COLOR_BGR2GRAY)
```

```python
_, th1_original = cv2.threshold(img, 240, 255, cv2.THRESH_BINARY) 
_, th2_original = cv2.threshold(img_denoised, 0, 255, cv2.THRESH_BINARY_INV) 
_, th3_1_original = cv2.threshold(img_denoised,86, 255, cv2.THRESH_BINARY_INV) 
_, th3_2_original = cv2.threshold(img_denoised,70, 255, cv2.THRESH_BINARY) 
th3_original = cv2.bitwise_and(th3_1, th3_2)

masks_raw = [th2_original, th3_original, th1_original]
```

```python
for raw, denoised in zip(masks_raw, masks_denoised_short):
    pred = denoised
    gt = raw
    
    tp = np.sum((pred == 255) & (gt == 255))
    print(tp)
    fp = np.sum((pred == 255) & (gt == 0))
    print(fp)
    fn = np.sum((pred == 0) & (gt == 255))
    print(fn)
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1_score = 2 * (precision * recall) / (precision + recall)

    # Print the F1-score
    print(f1_score)
```
